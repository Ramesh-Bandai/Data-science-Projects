# Building a Priliminary XGBoost model

```python
# Checking Balanced or imbalanced dataset
# % People left
sum(Y)/len(Y)
```

```python
#Splitting The Datsets train test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_encoded, Y, random_state=42,stratify=Y) #stratify=Y works as expected
```

```python
# Checking balnced or not
sum(y_train)/len(y_train),sum(y_test)/len(y_test)
```

```python
import xgboost as xgb
clf_xgb=xgb.XGBClassifier(objective='binary:logistic',missing=None,seed=42)  # missing=None -specified ur missing data '?' or 0
clf_xgb.fit(x_train,y_train,verbose=True,early_stopping_rounds=10,eval_metric='aucpr',eval_set=[(x_test,y_test)])
# seed=42 used for same result
#eval_metric 
#eval_set  based on testing dataset
```

```python
# Checking performance on testing dataset
#plotting Confusion Matrix for test dataset
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf_xgb, x_test, y_test,values_format='d',display_labels=['Did not Left','Left'])
```

# Optimizing Parameters using cross validation and GridSearch()

```python
# Round 1
param_grid={
            'max_depth':[3,4,5],
            'learning_rate':[0.1,0.01,0.05],
            'gamma':[0,0.25,1],
            'reg_lambda':[0,1.0,10.0],
            'scale_pos_weight':[1,3,5]
               }

from sklearn.model_selection import GridSearchCV
optimal_params=GridSearchCV(estimator=xgb.XGBClassifier(objective='binary:logistic',seed=42,subsample=0.9,colsample_bytree=0.5),
                            param_grid=param_grid,
                            scoring='roc_auc',
                            verbose=0,
                            n_jobs=10,
                            cv=3
                            )

optimal_params.fit(x_train,y_train,verbose=False,early_stopping_rounds=10,eval_metric='auc',eval_set=[(x_test,y_test)])
```

```python
# Checking Optimal Params_ for round 1
optimal_params.best_params_
```

```python
# Round 2
param_grid={
            'max_depth':[3], # from round 1
            'learning_rate':[0.1,0.5,1],
            'gamma':[0], #from 1
            'reg_lambda':[10,20,100],
            'scale_pos_weight':[3] # from 1
               }

from sklearn.model_selection import GridSearchCV
optimal_params=GridSearchCV(estimator=xgb.XGBClassifier(objective='binary:logistic',seed=42,subsample=0.9,colsample_bytree=0.5),
                            param_grid=param_grid,
                            scoring='roc_auc',
                            verbose=0,
                            n_jobs=10,
                            cv=3
                            )

optimal_params.fit(x_train,y_train,verbose=False,early_stopping_rounds=10,eval_metric='auc',eval_set=[(x_test,y_test)])
```

```python
# Checking Optimal Params_ for round 2
optimal_params.best_params_
```

# Final Model building

```python
import xgboost as xgb
clf_xgb=xgb.XGBClassifier(objective='binary:logistic',seed=42,gamma=0,learn_rate=0.1,max_depth=3,reg_lambda=10,scale_pos_weight=3,subsample=0.9,colsample_bytree=0.5) 
clf_xgb.fit(x_train,y_train,verbose=True,early_stopping_rounds=10,eval_metric='aucpr',eval_set=[(x_test,y_test)])
```

```python
# Checking performance on testing dataset
#plotting Confusion Matrix for test dataset
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf_xgb, x_test, y_test,values_format='d',display_labels=['Did not Left','Left'])
```
