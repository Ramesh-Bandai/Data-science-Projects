```python
# Finding Missing Rows
missing_df=df.loc[(df['EDUCATION']==0 )| (df['MARRIAGE']==0)] # OR oprator is important
missing_df
```

```python
print('No of missing values=',len(missing_df))
print('Percentage of Missing value=',len(missing_df)*100/len(df))
```

```python
final_df=df.loc[(df['EDUCATION']!=0 ) & (df['MARRIAGE']!=0)]  # & operator is important 
final_df
```

```python
len(final_df)
```

```python
#To downsample we split the dataframe into two dataframes,one for people who did not deafault and one for people who are default
df_no_default=final_df[final_df['DEFAULT']==0]
df_default=final_df[final_df['DEFAULT']==1]
```

```python
# checking No of No default and  default people
len(df_no_default),len(df_default)
```

```python
#downsample for no default people
from sklearn.utils import resample # sci-kit Learn resample  #Pandas dataframe also have resample choose correctly
df_no_default_downsampled=resample(df_no_default,replace=False,n_samples=1000,random_state=42) #for no default people
len(df_no_default_downsampled)
```

```python
#downsample default people
df_default_downsampled=resample(df_default,replace=False,n_samples=1000,random_state=42) #for default people
len(df_default_downsampled)
```

```python
# Combining two dataframes 
df_downsample=pd.concat([df_no_default_downsampled,df_default_downsampled]) # by default it combines rowwise
len(df_downsample)
```

```python
df_downsample
```

```python
# Splitting datset into X and Y
X=df_downsample.drop('DEFAULT',axis=1).copy()
X.head()
```

```python
Y=df_downsample['DEFAULT'].copy()
Y.head()
```

```python
X.columns
```

```python
# One hot encoding by using get dummies for categorical columns
X_encoded=pd.get_dummies(X,columns=['SEX', 'EDUCATION', 'MARRIAGE','PAY_0', 'PAY_2','PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'])

X_encoded.head()      
```

```python
# We split the data into training and testing and then scale them seperately to avoid data leakage.
# Data leakege occures when information about the training dataset influence the testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_encoded, Y, random_state=42)
from sklearn.preprocessing import scale
x_train_scaled=scale(x_train)
x_test_scaled=scale(x_test)
```

```python
# Building a Preliminary Support Vector Machine
from sklearn.svm import SVC
svm_clf=SVC(random_state=42)
svm_clf.fit(x_train_scaled,y_train)
```

```python
# we built SVM for classification.Let's check performance on testing dataset & draw confusion Matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(svm_clf, x_test_scaled, y_test,values_format='d',display_labels=['Did Not Default','Defaulted'])
plt.show()
```

```python
param_grid=[
    {    'C':[0.5,1,10,100], #values for C must be > 0
     'gamma':['scale',1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
    },
           ]
# We are including C=1 and gamma='scale' as a choices but they are default values

from sklearn.model_selection import GridSearchCV
optimal_params=GridSearchCV(SVC(),param_grid,cv=5,scoring='accuracy',verbose=0)
optimal_params.fit(x_train_scaled,y_train)
print(optimal_params.best_params_)
```

```python
#Building,Evaluting,Drawing & Interpreting the final Support vector Machine
# Now we have ideal value for C and gamma
svm_clf=SVC(random_state=42,C=100,gamma=0.001)
svm_clf.fit(x_train_scaled,y_train)
```

```python
# Let's draw another confusion Matrix to see the optimized support vector machine does better
plot_confusion_matrix(svm_clf, x_test_scaled, y_test,values_format='d',display_labels=['Did Not Default','Defaulted'])
plt.show()
```

```python
# Drawing a 'scree plot'
from sklearn.decomposition import PCA
pca=PCA()# by default PCA() centres the data but does not scale it.
pca_x_train=pca.fit_transform(x_train_scaled)
per_var=np.round(pca.explained_variance_ratio_*100,decimals=1)
fig= plt.figure(figsize=(15,8))
labels=[str(x) for x in range(1,len(per_var)+1)]
plt.bar(x=range(1,len(per_var)+1),height=per_var)
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel('Percentage of explained vriance')

plt.show()


```

```python
# Re-training and Re-optimizing the SVM 
train_pc1_coords=pca_x_train[:,0]
train_pc2_coords=pca_x_train[:,1]
#pc1 contains the x-axis coordinates of the data after PCA
#pc2 contains the y-axis coordinates of the data after PCA

# Now Center and scale the PCs....
pca_train_scaled=scale(np.column_stack((train_pc1_coords,train_pc2_coords)))

#Now we Optimize the SVM fit to the x and y-axis coordinates

param_grid=[
    {    'C':[1,10,100,1000], #values for C must be > 0
     'gamma':['scale',1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
    },
           ]
# We are including C=1 and gamma='scale' as a choices but they are default values

from sklearn.model_selection import GridSearchCV
optimal_params=GridSearchCV(SVC(),param_grid,cv=5,scoring='accuracy',verbose=0)
optimal_params.fit(pca_train_scaled,y_train)
print(optimal_params.best_params_)
```

```python
# Now we have optimal values for C and gamma let's draw the graph

svm_clf=SVC(random_state=42,C=1000,gamma=0.001)
svm_clf.fit(pca_train_scaled,y_train)

# Transform the test dataset with the PCA...
pca_x_test=pca.transform(x_train_scaled)
test_pc1_coords=pca_x_test[:,0]
test_pc2_coords=pca_x_test[:,1]

x_min=test_pc1_coords.min()-1
x_max=test_pc1_coords.max()+1
y_min=test_pc2_coords.min()-1
y_max=test_pc2_coords.max()+1

xx,yy=np.meshgrid(np.arange(start=x_min,stop=x_max,step=0.1),
                  np.arange(start=y_min,stop=y_max,step=0.1))

Z=svm_clf.predict(np.column_stack((xx.ravel(),yy.ravel())))
Z=Z.reshape(xx.shape)

fig,ax=plt.subplots(figsize=(12,10))
ax.contourf(xx,yy,Z,alpha=0.1)

import matplotlib.colors as colors
cmap=colors.ListedColormap(['#e41a1c','#4daf4a'])
scatter=ax.scatter(test_pc1_coords,test_pc2_coords,c=y_train,cmap=cmap,s=100,edgecolors='k',alpha=0.7)
# Now create a Legend
legend=ax.legend(scatter.legend_elements()[0],
                 scatter.legend_elements()[1],
                 loc='upper right')
legend.get_texts()[0].set_text('No Default')
legend.get_texts()[1].set_text('Yes Default')

# Adding axis lebels and title's 
ax.set_ylabel('PC2')
ax.set_xlabel('PC1')
ax.set_title('Decision Surface using the PCA transformed/projected features')
plt.show()
```
